```python
import requests
```

```python
address = "bc1qmcvv20xmujhhelyas5r59gkgvl92dhdwr9v4t4"
utxos = requests.get(f"https://blockchain.info/unspent?active={address}").json()["unspent_outputs"]

with open(f"unspent_{address}.txt", "w") as f:
    f.write("TX_ID\tOutput\tAmount\n")

    for utxo in utxos:
        f.write(
            f"{utxo['tx_hash_big_endian']}\t"
            f"{utxo['tx_output_n']}\t"
            f"{utxo['value']}\n"
        )

balance = requests.get(
    f"https://blockchain.info/balance?active={address}").json()

with open(f"balance_{address}.txt", "w") as f:
    f.write(str(balance[address]))
```
